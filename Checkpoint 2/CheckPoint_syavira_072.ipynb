### **Outliers Values**

Dari hasil pengecekan nilai ekstrem (outlier) pada kolom bertipe numerik berikut, dapat terlihat kolom:
*   **product_id**

    Dengan presentase sekitar 12%

*   **beauty_point_earned**

    Dengan presentase sekitar 4%

*   **average_rating**

    Dengan presentase sekitar 24%

*   **total_reviews**

    Dengan presentase sekitar 15%

*   **total_recommended_count**

    Dengan presentase sekitar 15%

*   **total_repurchase_maybe_count**

    Dengan presentase sekitar 15%

*   **total_repurchase_no_count**

    Dengan presentase sekitar 14%

*   **total_repurchase_yes_count**

    Dengan presentase sekitar 15%

*   **total_in_wishlist**

    Dengan presentase sekitar 11%    

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## **Comparison/Perbandingan**

Aktivitas: Membandingkan brand (brand_name) dengan total review dari pelanggan (total_reviews).

Tujuan: Untuk mengetahui 10 brand teratas yang paling populer.

Visualisasi: Bar Chart (Grafik Batang).

```python
popular_brand = df.groupby('brand_name')['total_reviews'].sum().sort_values(ascending=False)

top_brand = popular_brand.head(10)

plt.figure(figsize=(10,6))
sns.barplot(
    x=top_brand.index,
    y=top_brand.values,
    palette='viridis',
    hue=top_brand.index,
    legend=False
)

plt.title('Top 10 Brand Paling Populer Berdasarkan Total Reviews')
plt.xlabel('Brand')
plt.ylabel('Total Reviews')
plt.xticks(rotation=45)
plt.show()
```

## **Relationship/Hubungan**

Aktivitas: Menganalisis hubungan antara nilai rating rata-rata produk (average_rating) dengan jumlah rekomendasi pelanggan (total_recommended_count).

Tujuan: Melihat apakah rating tinggi berhubungan dengan banyaknya orang yang merekomendasikan produk.

Visualisasi: Heatmap

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['average_rating', 'total_recommended_count']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('Korelasi antara Average Rating dan Total Recommended Count')
plt.show()
```

Dari hasil visualisasi tersebut dapat disimpulkan:

Heatmap tersebut menunjukkan bahwa hubungan antara average_rating dan total_recommended_count sangat lemah, dengan korelasi hanya sekitar 0.12.

Artinya, meskipun sebuah produk memiliki rating yang tinggi, hal itu tidak selalu berarti produk tersebut banyak direkomendasikan oleh pelanggan. Jumlah rekomendasi tampaknya dipengaruhi oleh faktor lain seperti popularitas brand, jumlah pengguna yang aktif memberikan review, atau tingkat penggunaan produk, sehingga rating tinggi tidak otomatis berbanding lurus dengan banyaknya rekomendasi.
