## **Business Objective**

Dataset Sociolla: All Brands Products Catalog

Dataset ini mencakup informasi tentang lebih dari 7.500 produk skincare dan make up yang diambil dari Sociolla. Tujuan dari analisis ini adalah untuk memahami brand skincare atau make up mana yang terlihat lebih populer dan lebih banyak menarik perhatian pengguna. Dengan mengetahui brand mana yang paling menonjol, suatu bisnis dapat memahami tren pasar, melihat preferensi konsumen, dan mengetahui brand mana yang memiliki potensi kerja sama lebih besar.
